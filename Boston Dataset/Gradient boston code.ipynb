## Required Imports.

```python
import numpy as np
from sklearn import model_selection
from sklearn import preprocessing
```

## Gradient Descent algorithm.

```python
def stepGradient(X,Y,learning_rate,num_iterations,coeff):
    m_slope=np.zeros(len(X[0]))
    data_points=len(X)
    features=len(X[0])
    for i in range(data_points):
        x=X[i]
        y=Y[i]
        for j in range(features):
            m_slope[j]+=(-2/data_points)*(y-sum(coeff*x))*x[j]
    next_m=coeff-(learning_rate*m_slope)
    return next_m

def cost(X,Y,m):
    total_cost=0
    for i in range(len(X)):
        total_cost+=((1/len(X))*(Y[i]-sum(m*X[i]))**2)
    return total_cost
            
def GD(X,Y):
    learning_rate=0.23
    num_iterations=500
    X=np.append(X,np.ones(len(X)).reshape(-1,1),axis=1)
    coeff=np.zeros(len(X[0]))
    for i in range(num_iterations):
        coeff=stepGradient(X,Y,learning_rate,num_iterations,coeff)
        print('Cost-',str(i+1),' ',cost(X,Y,coeff))
    return coeff
```

## Loading the training data and applying gradient descent algorithm.

```python
data=np.genfromtxt('training_boston.csv',delimiter=',')
X_train=data[:,:-1]
Y_train=data[:,-1]
add_features=[]
for i in X_train:
    add_features.append(i**2)
add_features=np.array(add_features)
X_train=np.append(X_train,add_features,axis=1)
scaler=preprocessing.MinMaxScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
m=GD(X_train,Y_train)
print(m)
```

## Loading the testing data.

```python
test_data=np.genfromtxt('test_boston.csv',delimiter=',')
X_test=test_data
test_feature_add=[]
for i in X_test:
    test_feature_add.append(i**2)
test_feature_add=np.array(test_feature_add)
X_test=np.append(X_test,test_feature_add,axis=1)
scaler.fit(X_test)
X_test=scaler.transform(X_test)
X_test=np.append(X_test,np.ones(len(X_test)).reshape(-1,1),axis=1)
```

## Finding the predictions and saving it in file. 

```python
Y_pred=[]
for i in X_test:
    y=sum(m*i)
    Y_pred.append(y)
Y_pred=np.array(Y_pred)
np.savetxt('predictions1.csv',Y_pred,fmt="%.5f", delimiter=",")
```
