# Importing modules.

```python
import numpy as np
from sklearn import preprocessing
```

# Gradient Descent Algorithm.

```python
def step_gradient(X,Y,learning_rate,coeff):
    m_slope=np.zeros(len(X[0]))
    features=len(X[0])
    data_points=len(X)
    for i in range(data_points):
        x=X[i]
        y=Y[i]
        for j in range(features):
            m_slope[j]+=(-2/data_points)*(y-sum(coeff*x))*x[j]
    new_m=coeff-(learning_rate*m_slope)
    return new_m

def cost(X,Y,coeff):
    total_cost=0
    for i in range(len(X)):
        y=Y[i]
        total_cost+=(1/len(X))*((y-sum(coeff*X[i]))**2)
    return total_cost
        
def gd(X,Y,learning_rate,iterations):
    coeff=np.zeros(len(X[0]))
    for i in range(iterations):
        coeff=step_gradient(X,Y,learning_rate,coeff)
        print('Cost'+str(i+1)+'->',cost(X,Y,coeff))
    return coeff

def Gradient_descent(X,Y):
    learning_rate=0.2
    iterations=1000
    X=np.append(X,np.ones(len(X)).reshape(-1,1),axis=1)
    m=gd(X,Y,learning_rate,iterations)
    return m
    
    
```

# Loading Training data.

```python
Train_data=np.genfromtxt('training.csv',delimiter=',')
X_train=Train_data[:,:-1]
Y_train=Train_data[:,-1]

square = []
for i in X_train:
    square.append(i**2)
square = np.array(square)
X_train = np.append(X_train, square, axis = 1)

scaler=preprocessing.StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
```

#  Using Gradient Descent Algo on training data.

```python
m=Gradient_descent(X_train,Y_train)

```

# Loading testing data.

```python
X_test=np.genfromtxt('test.csv',delimiter=',')

sq_test=[]

for i in X_test:
    sq_test.append(i**2)

sq_test=np.array(sq_test)

X_test=np.append(X_test,sq_test,axis=1)


scaler.fit(X_test)
X_test=scaler.transform(X_test)


X_test=np.append(X_test,np.ones(len(X_test)).reshape(-1,1),axis=1)
```

# Predictions for X_test and saving in csv file.

```python

pred=[]

for i in X_test:
    pred.append(sum(i*m))
    
Y_pred=np.array(pred)


np.savetxt('predictions2.csv',Y_pred,fmt='%.5f',delimiter=',')
```
